```python
import pandas as pd

df = pd.read_csv("HR_Attrition.csv")
print("Shape:", df.shape)
df.head()
```

```python
print(df.columns.tolist())
```

```python
print("Attrition Count:")
print(df['Attrition'].value_counts())

print("\nAttrition Percentage:")
print(df['Attrition'].value_counts(normalize=True) * 100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
colors = ['steelblue', 'coral']
df['Attrition'].value_counts().plot(kind='bar', color=colors)

plt.title('Employee Attrition Count', fontsize=16, fontweight='bold')
plt.xlabel('Attrition', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('attrition_count.png')
plt.show()

print("Chart saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Department', hue='Attrition', palette=['steelblue', 'coral'])

plt.title('Attrition by Department', fontsize=16, fontweight='bold')
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend(title='Attrition')
plt.tight_layout()
plt.savefig('attrition_by_department.png')
plt.show()

print("Chart saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20, palette=['steelblue', 'coral'])

plt.title('Age Distribution by Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.tight_layout()
plt.savefig('age_attrition.png')
plt.show()

print("Chart saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette=['steelblue', 'coral'])

plt.title('Monthly Income by Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Attrition', fontsize=12)
plt.ylabel('Monthly Income', fontsize=12)
plt.tight_layout()
plt.savefig('income_attrition.png')
plt.show()

print("Chart saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette=['steelblue', 'coral'])

plt.title('Overtime vs Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Overtime', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend(title='Attrition')
plt.tight_layout()
plt.savefig('overtime_attrition.png')
plt.show()

print("Chart saved!")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition', palette=['steelblue', 'coral'])

plt.title('Job Satisfaction vs Attrition', fontsize=16, fontweight='bold')
plt.xlabel('Job Satisfaction (1=Low, 4=High)', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend(title='Attrition')
plt.tight_layout()
plt.savefig('jobsatisfaction_attrition.png')
plt.show()

print("Chart saved!")
```
